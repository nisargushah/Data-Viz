```
countries = list(df.index)
```

```
countries
```

```
max_infection_rate = []
for c in countries:
    max_infection_rate.append(df.loc[c].diff().max())
df["max_infection_rate"] = max_infection_rate
```

```
df["max_infection_rate"]
```

```
corona_data = pd.DataFrame(df["max_infection_rate"])
corona_data.head()
```

```
happiness_report= pd.read_csv("worldwide_happiness_report.csv")
happiness_report
```
